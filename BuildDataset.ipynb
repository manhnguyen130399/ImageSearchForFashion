```python
y_train = dataset_train["label"]
y_train = y_train.to_numpy()
y_train
```

```python
x_train = dataset_train.drop('label', axis=1)
x_train = x_train.to_numpy()
x_train = np.array(x_train,dtype=np.uint8)
x_train
```

```python
# đảo ảnh
list_pixel =[]
column=['label']
for i in range(1,785):
  str1 = 'pixel'+str(i)
  column.append(str1)
for i in range(0, len(dataset_train)):
  m3 = cv2.flip(x_train[i].reshape(28,28),1)
  m3 = m3.astype(np.uint8)
  list_pixel.append(np.insert(m3.flatten(),0,y_train[i]))
```

```python
df_img = pd.DataFrame(columns=column)
for i in range(0, len(list_pixel)):
  df_img.loc[i]= list_pixel[i]
# append data train
# df_img.append(dataset_train)
df_img.to_csv("GenerateData.csv")
```
